To compute the Fourier series for the given triangular wave, we need to determine the coefficients for the cosine and sine modes. The Fourier series representation of a periodic function f(x) with period 2L is given by:
$
f(x) = a_0/2 + \sum_n(a_n cos(n\pi x/L) + b_n sin(n\pi x/L))
$

where $a_0$, $a_n$, and $b_n$ are the Fourier coefficients, and $n$ is the mode number. In this case, we have a triangular wave with period 2, so L = 1.

To compute the Fourier coefficients, we can use the following formulas:

$$
\begin{align}

a_0 &= \frac{1}{L} \int_{-L}^{L} f(x) dx \\
a_n &= \frac{1}{L} \int_{-L}^{L} f(x) cos(n\pi x/L) dx \\
b_n &= \frac{1}{L} \int_{-L}^{L} f(x) sin(n\pi x/L) dx \\
\end{align}
$$

In this case, L = 1, and the integral limits are from -1 to 1. Let's compute the coefficients for the first 100 cosine and sine modes.

$a_0$:

$$
\begin{align}
a_0 &= \frac{1}{1} \int_{-1}^{1} f(x) dx \\
&= \int_{-1}^{1} (1 - |x|) dx \\
&= 2\int_{0}^{1} (1 - x) dx \\
&= 2\left[x - \frac{x^2}{2}\right] \quad \text{evaluated from 0 to 1} \\
&= 2(1 - \frac{1}{2} - 0 + 0) \\
&= 1 \\
\end{align}
$$



$a_n$:

$$

a_n &= \frac{1}{1} \int_{-1}^{1} f(x) cos(n\pi x) dx \\

\end{align}
$$


To compute this integral, we need notice that the function is even:

$$

# Exercise 2-3

Use the FFT to solve the Kortewegde Vries (KdV) equation,

$$
\frac{\partial u}{\partial t} + \frac{\partial^3 u}{\partial x^3} - u \frac{\partial u}{\partial x} = 0
$$

on a large domain with an initial condition $u(x, 0) = sech(x)$. Plot the evolution.

```python
L = 20     # Length of domain
N = 2000   # Number of discretization points
dx = L/N
x = np.arange(-L/2,L/2,dx) # Define x domain

# Define discrete wavenumbers
kappa = 2 * np.pi * np.fft.fftfreq(N, d=dx)

# Initial condition
u0 = 1/np.cosh(x)
u0_hat = np.fft.fft(u0)

# Time parameters
dt = 0.0025
t_start = 0
t_end = 2.5  # Adjusted for practical computation time
num_time_steps = int((t_end - t_start) / dt)
t_eval = np.linspace(t_start, t_end, num_time_steps + 1)

# Precompute linear operator and integrating factors
L_operator = -1j * kappa ** 3
E = np.exp(L_operator * dt)
E2 = np.exp(L_operator * dt / 2)
```

```python
# Precompute ETDRK4 coefficients
M = 16  # Number of points for contour integral
r = np.exp(1j * np.pi * (np.arange(1, M + 1) - 0.5) / M)
L_dt = L_operator[:, None] * dt + r[None, :]
Q = dt * np.mean((np.exp(L_dt / 2) - 1) / (L_dt / 2), axis=1)
f1 = dt * np.mean((-4 - L_dt + np.exp(L_dt) * (4 - 3 * L_dt + L_dt ** 2)) / (L_dt ** 3), axis=1)
f2 = dt * np.mean((2 + L_dt + np.exp(L_dt) * (-2 + L_dt)) / (L_dt ** 3), axis=1)
f3 = dt * np.mean((-4 - 3 * L_dt - L_dt ** 2 + np.exp(L_dt) * (4 - L_dt)) / (L_dt ** 3), axis=1)

# Initialize solution storage
u_hat = u0_hat.copy()
u_list = [u0.copy()]

# Time-stepping loop
for n in range(num_time_steps):
    t = n * dt
    # Compute nonlinear term N(u)
    u = np.fft.ifft(u_hat)
    N0 = -0.5j * kappa * np.fft.fft(u * u)
    
    # First stage
    a_hat = E2 * u_hat + Q * N0
    a = np.fft.ifft(a_hat)
    Na = -0.5j * kappa * np.fft.fft(a * a)
    
    # Second stage
    b_hat = E2 * u_hat + Q * Na
    b = np.fft.ifft(b_hat)
    Nb = -0.5j * kappa * np.fft.fft(b * b)
    
    # Third * a_hat + Q * (2 * Nb - N0)
    c = np.fft.ifft(c_hat)
    Nc = -0.5j * kappa * np.fft.fft(c * c)
    
    # Update u_hat
    u_hat = E * u_hat + N0 * f1 + 2 * (Na + Nb) * f2 + Nc * f3
    if n % 10 == 0:
        u_list.append(np.real(np.fft.ifft(u_hat)))

# Prepare data for plotting
U = np.array(u_list)
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Subsample the solution for plotting
u_plot = U[0:-1:10,:]
time_indices = np.arange(len(u_plot))

for j, u_slice in enumerate(u_plot):
    ys = np.full_like(x, time_indices[j] * dt * 10)
    ax.plot(x, ys, u_slice)

ax.set_xlabel('x')
ax.set_ylabel('Time')
ax.set_zlabel('u(x, t)')
plt.title('Evolution of u(x, t) over time')
plt.show()
```

# Exercise 2-4

Use the FFT to solve the KuramotoSivashinsky (KS) equation

$$
\frac{\partial u}{\partial t} + \frac{\partial^2 u}{\partial x^2} + \frac{\partial^4 u}{\partial x^4} + \frac{1}{2}\frac{\partial u^2}{\partial x} = 0
$$

on a large domain with an initial condition $u(x, 0) = sech(x)$. Plot the evolution.

```python
L = 20     # Length of domain
N = 1000   # Number of discretization points
dx = L/N
x = np.arange(-L/2,L/2,dx) # Define x domain

# Define discrete wavenumbers
kappa = 2 * np.pi * np.fft.fftfreq(N, d=dx)

# Initial condition
u0 = 1/np.cosh(x)

# Simulate PDE in spatial domain
dt = 0.025
t = np.arange(0,100*dt,dt)
```

```python
def KS(u,t):
    uhat = np.fft.fft(u)

    d_uhat = (1j)*kappa*uhat
    dd_uhat = np.power((1j)*kappa,2) * uhat
    dddd_uhat = np.power((1j)*kappa,4) * uhat

    d_u = np.fft.ifft(d_uhat)
    dd_u = np.fft.ifft(dd_uhat)
    dddd_u = np.fft.ifft(dddd_uhat)
    
    du_dt = -dd_u - dddd_u - 1/2*d_u**2
    return du_dt.real
```

```python
u = sp.integrate.odeint(KS,u0,t)

# Waterfall plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

u_plot = u[0:-1:10,:]
for j in range(u_plot.shape[0]):
    ys = j*np.ones(u_plot.shape[1])
    ax.plot(x,ys,u_plot[j,:])
```

Now, compute the solution to the 2D heat equation on a circular

and [1.5, 1.5] in $y$. You can impose the boundary conditions by enforcing the

Simulate the unsteady heat equation for the following boundary conditions:

Include your code and show some plots of your solutions to the heat equation.
Plot the temperature distribution for each case (1) early on in the diffusion process, (2) near steady state, and (3) somewhere in the middle.

```python
# Define grid parameters
N = 300
x_min, x_max = -1.5, 1.5
y_min, y_max = -1.5, 1.5
dx = (x_max - x_min) / (N - 1)
dy = (y_max - y_min) / (N - 1)
x = np.linspace(x_min, x_max, N)
y = np.linspace(y_min, y_max, N)
X, Y = np.meshgrid(x, y)

# Define circle parameters
a = 1
r = 1

# Define the disk
disk = X**2 + Y**2 <= r**2

# Stability condition for explicit scheme
dt = 0.2 * dx**2 / (4 * a**2)
nt = 10000
```

```python
def apply_boundary_conditions_a(u):
    theta = np.arctan2(Y, X)
    boundary = (np.abs(X**2 + Y**2 - r**2) <= (dx + dy))
    left_half = boundary & (np.cos(theta) < 0)
    right_half = boundary & (np.cos(theta) >= 0)
    u[left_half] = 1.0
    u[right_half] = 2.0
    return u
```

```python
def apply_boundary_conditions_b(u):
    theta = np.arctan2(Y, X)
    boundary = (np.abs(X**2 + Y**2 - r**2) <= (dx + dy))
    u[boundary] = np.cos(2*np.pi*theta[boundary])
    return u
```

```python
def enforce_disk(u):
    u[~disk] = 0.0
    return u
```

```python
def plot_temperature(u, timestep):
    plt.figure(figsize=(6,6))
    plt.contourf(X, Y, u, levels=50, cmap='inferno')
    plt.colorbar(label='Temperature')
    plt.title(f'Temperature at timestep {timestep}')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()
```

```python
def compute_laplacian(u):
    laplacian = np.zeros_like(u)
    laplacian[1:-1, 1:-1] = (u[2:, 1:-1] - 2 * u[1:-1, 1:-1] + u[:-2, 1:-1]) / dx**2 + \
                            (u[1:-1, 2:] - 2 * u[1:-1, 1:-1] + u[1:-1, :-2]) / dy**2
    return laplacian
```

```python
def simulate_heat_equation(u, boundary_func, nt):
    data = []
    for n in range(nt):
        laplacian = compute_laplacian(u)
        u[disk] += dt * a**2 * laplacian[disk]  # Update only inside the disk
        u = boundary_func(u)  # Apply boundary conditions
        u = enforce_disk(u)   # Set temperature outside the disk to 0
        data.append(u.copy())
    return data
```

\end{align}

```python
# b)

fft_rush = fft(rush)
n = len(rush)

# Compute Power Spectral Density
psd_rush = np.abs(fft_rush)**2

# Plot PSD
plt.figure(figsize=(10, 6))
plt.plot(np.linspace(0, FS, n), psd_rush)
plt.title('Power Spectral Density of Original Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
plt.grid(True)
plt.show()

# Plot spectrogram
frequencies, times, Sxx = spectrogram(rush, FS)
plt.figure(figsize=(10, 6))
plt.pcolormesh(times, frequencies, 10 * np.log10(Sxx))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.title('Spectrogram of Original Signal')
plt.colorbar(label='Power/Frequency (dB/Hz)')
plt.show()
```

```python
# c)
data_noisy = loadmat(os.path.join("..", "DATA", "r2112noisy.mat"))
rushnoisy = data_noisy['rushnoisy'].flatten()

Audio(rushnoisy, rate=FS) # Loud noise
```

```python
fft_rushnoisy = fft(rushnoisy)
n_noisy = len(rushnoisy)

# Zero out last three-quarters of Fourier components
fft_rushnoisy[int(n_noisy/4):] = 0

# Inverse FFT to reconstruct the cleaned signal
cleaned_signal = 2 * np.real(ifft(fft_rushnoisy))
```

```python
# Power Spectral Density for the noisy signal
psd_noisy = np.abs(fft_rushnoisy)**2

# Plot PSD before and after cleaning
plt.figure(figsize=(10, 6))
plt.subplot(2, 1, 1)
plt.plot(np.linspace(0, FS, n_noisy), np.abs(fft(rushnoisy))**2)
plt.title('Power Spectral Density of Noisy Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

plt.subplot(2, 1, 2)
plt.plot(np.linspace(0, FS, n_noisy), psd_noisy)
plt.title('Power Spectral Density of Cleaned Signal')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')

plt.tight_layout()
plt.show()
```

```python
# Spectrograms of the noisy and cleaned signal
frequencies_noisy, times_noisy, Sxx_noisy = spectrogram(rushnoisy, FS)
plt.figure(figsize=(10, 6))
plt.pcolormesh(times_noisy, frequencies_noisy, 10 * np.log10(Sxx_noisy))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.title('Spectrogram of Noisy Signal')
plt.colorbar(label='Power/Frequency (dB/Hz)')
plt.show()

frequencies_clean, times_clean, Sxx_clean = spectrogram(cleaned_signal, FS)
plt.figure(figsize=(10, 6))
plt.pcolormesh(times_clean, frequencies_clean, 10 * np.log10(Sxx_clean))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [s]')
plt.title('Spectrogram of Cleaned Signal')
plt.colorbar(label='Power/Frequency (dB/Hz)')
plt.show()
```

# Exercise 2-9
The convolution integral and the impulse response may be used to simulate how an audio signal would sound under various conditions, such as in a long hallway, in a concert hall, or in a swimming pool.
The basic idea is that you can record the audio response to an impulsive sound in a given location, like a concert hall. For example, imagine that you put a microphone in the most expensive seats in the hall and then record the sound from a shotgun blast up on the stage. (Do not try this!!) Then, if you have a flat studio recording of some other audio, you can simulate how it would have sounded in the concert hall by convolving the two signals. Download and unzip sounds.zip to find various sounds and impulse-response filters. Convolve the various audio files (labeled sound1.wav, ...) with the various filters (labeled FilterXYZ.wav, ...). In MATLAB, use the wavread command to load and the conv command to convolve. It is best to add 10% of the filtered audio (also known as wet audio) to 90% of the original audio (also known as dry audio). Listen to the filtered audio, as well as the original audio and the impulse-response filters (note that each sound has a sampling rate of FS=11,025). However, you will need to be careful when adding the 10% filtered and 90% unfiltered signals, since the filtered audio will not necessarily have the same length as the filtered audio. There is a great video explaining how to actually create these impulse responses:
http://www.audioease.com/Pages/Altiverb/sampling.php

```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# Load the original sound and filter
FS, sound = wavfile.read(os.path.join("..", "DATA","sounds", "sound3.wav"))
_, impulse_response = wavfile.read(os.path.join("..", "DATA","sounds", "FilterHall.wav"))

# Check that both files have the same sampling rate
if FS != 11025:
    raise ValueError("Sampling rate mismatch. Expected FS=11,025.")

# Normalize audio signals to avoid clipping
sound = sound / np.max(np.abs(sound))
impulse_response = impulse_response / np.max(np.abs(impulse_response))

impulse_response = np.mean(impulse_response, axis=1)
```

```python
# Convolve the sound with the impulse response
filtered_audio = np.convolve(sound, impulse_response, mode='full')

# Normalize the filtered audio to avoid clipping
filtered_audio = filtered_audio / np.max(np.abs(filtered_audio))
```

```python
# Make sure the lengths match by truncating the filtered audio if necessary
min_length = min(len(sound), len(filtered_audio))
sound = sound[:min_length]
filtered_audio = filtered_audio[:min_length]

# Combine 90% dry and 10% wet audio
combined_audio = 0.5 * sound + 0.5 * filtered_audio

# Normalize the combined audio
combined_audio = combined_audio / np.max(np.abs(combined_audio))
```

```python
from scipy.io.wavfile import write

# Save the combined audio
write("combined_audio.wav", FS, combined_audio)
```

```python
import sounddevice as sd

# Play the combined audio
sd.play(combined_audio, FS)
```

# Exercise 2-10
Verify the following properties of the Laplace transform

### a) Exponetial: $L\{e^{\lambda t}\} = \frac{1}{s-\lambda}$

$$
\begin{align}
L\{e^{\lambda t}\} &= \int_{0}^{\infty} e^{-st} e^{\lambda t} dt \\
&= \int_{0}^{\infty} e^{-(s-\lambda)t} dt \\
&= \frac{1}{s-\lambda} \left[-e^{-(s-\lambda)t}\right]_{0}^{\infty} \\
&= \frac{1}{s-\lambda} \left[0 - (-1)\right] \\
&= \frac{1}{s-\lambda}
\end{align}
$$


### b) Linearity: $L\{af(t) + bg(t)\} = aL\{f(t)\} + bL\{g(t)\}$

$$
\begin{align}
L\{af(t) + bg(t)\} &= \int_{0}^{\infty} e^{-st} (af(t) + bg(t)) dt \\
&= \int_{0}^{\infty} e^{-st} af(t) dt + \int_{0}^{\infty} e^{-st} bg(t) dt \\
&= a\int_{0}^{\infty} e^{-st} f(t) dt + b\int_{0}^{\infty} e^{-st} g(t) dt \\
&= aL\{f(t)\} + bL\{g(t)\}
\end{align}
$$

### c) Convolution: $L\{f(t) * g(t)\} = L\{f(t)\}L\{g(t)\}$

$$
\begin{align}
L\{f(t) * g(t)\} &= \int_{0}^{\infty} e^{-st} \int_{0}^{t} f(\tau) g(t-\tau) d\tau dt \\
&= \int_{0}^{\infty} \int_{0}^{t} e^{-st} f(\tau) g(t-\tau) d\tau dt \\
&= \int_{0}^{\infty}\left(\int_{0}^{t}  f(\tau) g(t-\tau) d\tau\right)e^{-st} dt \\
&= \int_{0}^{\infty}\left(\int_{0}^{\infty}  f(\tau) g(t-\tau) d\tau\right)e^{-st} dt \\
&= \int_{0}^{\infty} f(\tau) \left(\int_{0}^{\infty}  g(t-\tau) e^{-st} dt\right) d\tau \\
\end{align}
$$

let make the substitution $u = t - \tau$ and $du = dt$. The bounds of integration are from $-u$ to $\infty$. However since we want $g(\tau)$ to vanish for negative values of $\tau$, we can change the bounds of integration to $0$ to $\infty$.

$$
\begin{align}
L\{f(t) * g(t)\} &= \int_{0}^{\infty} f(\tau) \left(\int_{0}^{\infty}  g(u) e^{-s(u+\tau)} du\right) d\tau \\
&= \int_{0}^{\infty} f(\tau) \left(\int_{0}^{\infty}  g(u) e^{-su} e^{-s\tau} du\right) d\tau \\
&= \int_{0}^{\infty} f(\tau) e^{-s\tau} \left(\int_{0}^{\infty}  g(u) e^{-su} du\right) d\tau \\
&= \int_{0}^{\infty} f(\tau) e^{-s\tau} L\{g(t)\} d\tau \\
&= L\{f(t)\}L\{g(t)\}
\end{align}
$$


### d) Constant: $L\{1\} = \frac{1}{s}$

$$
\begin{align}
L\{1\} &= \int_{0}^{\infty} e^{-st} dt \\
&= \left[-\frac{e^{-st}}{s}\right]_{0}^{\infty} \\
&= \frac{1}{s} \left[0 - (-1)\right] \\
&= \frac{1}{s}
\end{align}
$$


### e) Delta function: $L\{\delta(t)\} = 1$

$$
\begin{align}
L\{\delta(t)\} &= \int_{0}^{\infty} e^{-st} \delta(t) dt \\
&= e^{-s\cdot 0}\\
&= 1
\end{align}
$$

# Exercise 2-11
Use the Laplace transform to solve for the general solution to the following differential equations:

$$
\begin{align}
x''+ax'+bx &= u(t)
\end{align}
$$

Taking the laplace transform of both sides:

$$
\begin{align}
L\{x''+ax'+bx\} &= L\{u(t)\} \\
L\{x''\} + aL\{x'\} + bL\{x\} &= L\{u(t)\} \\
s^2L\{x\} - sx(0) - x'(0) + asL\{x\} - ax(0) + bL\{x\} &= L\{u(t)\} \\
(s^2 + as + b)L\{x\} - sx(0) - x'(0) - ax(0) &= L\{u(t)\} \\
(s^2 + as + b)L\{x\} &= L\{u(t)\} + sx(0) + x'(0) + ax(0) \\
L\{x\} &= \frac{L\{u(t)\} + sx(0) + x'(0) + ax(0)}{s^2 + as + b} \\
\end{align}
$$

### What if $u(t)=0$?

$$
\begin{align}
L\{x\} &= \frac{L\{0\} + sx(0) + x'(0) + ax(0)}{s^2 + as + b} \\
&= \frac{sx(0) + x'(0) + ax(0)}{s^2 + as + b} \\
\end{align}
$$

### What if $u(t) = \delta(t)$?

$$
\begin{align}
L\{x\} &= \frac{L\{\delta(t)\} + sx(0) + x'(0) + ax(0)}{s^2 + as + b} \\
&= \frac{1 + sx(0) + x'(0) + ax(0)}{s^2 + as + b} \\
\end{align}
$$

### What if $u(t) = 1$?

\end{align}
